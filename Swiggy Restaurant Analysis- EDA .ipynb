# EDA

```python
df5=df10['City'].value_counts().head(10)
fig=px.bar(df5)
fig.show()
fig.write_image("Cities.png")
```

```python
df6=df10[df10['City']=='Bangalore']['Area'].value_counts().head(10)
fig=px.bar(df6)
fig.show()
fig.write_image("City(Banglore).png")
```

```python
plt.figure(figsize=(10,10))
fig=px.scatter(df10,x='Price',y='Avg ratings',color='City')
fig.show()
fig.write_image("price vs ratings.png")
```

```python
df1=df10.groupby('City')['Delivery time'].mean().sort_values()
fig=px.bar(df1)
fig.show()
fig.write_image("City vs Delivery Time.png")
```

```python
df1=df10.groupby('Area')['Delivery time'].mean().sort_values()
fig=px.bar(df1)
fig.show()
fig.write_image("Area vs Delivery Time.png")
```

```python
import plotly.graph_objects as go
correlation_matrix=df10.corr()
fig=go.Figure(data=go.Heatmap(z=correlation_matrix,x=correlation_matrix.columns,y=correlation_matrix.columns))
fig.show()
fig.write_image("Correlation.png")
```

```python
df6=df10.groupby('Area')['Price'].mean().sort_values().head(15)
fig=px.bar(df6)
fig.show()
fig.write_image("Area vs Price.png")
```

```python
ratings_higherthan45 = df10[df10['Avg ratings'] > 4.5]
ratings_lowerthan45 = df10[df10['Avg ratings'] <= 4.5]

perce_above45 = len(ratings_higherthan45) / len(df10) * 100
perce_below45 = 100 - perce_above45

rating_df = pd.DataFrame({
    'Rating': ['Above 4.5', '4.5 and Below'],
    'Percentage': [perce_above45, perce_below45]
})

fig = px.pie(rating_df, names='Rating', values='Percentage',
             title='Distribution of Ratings')
fig.show()
fig.write_image("Distribution of ratings.png")
```
